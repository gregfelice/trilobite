**Model function**

As described in lecture, the model function for linear regression (which is a function that maps from x to y) is represented as

,(())=()+(1)

The formula above is how you can represent straight lines - different values of 
and 

give you different straight lines on the plot.

Let's try to get a better intuition for this through the code blocks below. Let's start with =100
and =100.

Note: You can come back to this cell to adjust the model's w and b parameters

```python
w = 200 # weight
b = 100 # bias
print(f"w: {w}")
print(f"b: {b}")
```

Now, let's compute the value of ,(()) for your two data points. You can explicitly write this out for each data point as -

for (0), f_wb = w * x[0] + b

for (1), f_wb = w * x[1] + b

For a large number of data points, this can get unwieldy and repetitive. So instead, you can calculate the function output in a for loop as shown in the compute_model_output function below

* Note: The argument description (ndarray (m,)) describes a Numpy n-dimensional array of shape (m,). (scalar) describes an argument without dimensions, just a magnitude.
* Note: np.zero(n) will return a one-dimensional numpy array with  entries 

```python
def compute_model_output(x, w, b):
    """
    Computes the prediction of a linear model
    Args:
      x (ndarray (m,)): Data, m examples 
      w,b (scalar)    : model parameters  
    Returns
      f_wb (ndarray (m,)): model prediction
    """
    m = x.shape[0] # the dimensional structure of the array
    f_wb = np.zeros(m) # prepares an array to accept y hat
    for i in range(m):
        f_wb[i] = w * x[i] + b # linear regression formula, accepting feature value (x), weight and bias of model
        # y = mx + b
        
    return f_wb
```

Now let's call the `compute_model_output` function and plot the output..

```python
tmp_f_wb = compute_model_output(x_train, w, b,) # pass in the x_train array

# Plot our model prediction
plt.plot(x_train, tmp_f_wb, c='b',label='Our Prediction') # the y hat values

# Plot the data points
plt.scatter(x_train, y_train, marker='x', c='r',label='Actual Values')

# Set the title
plt.title("Housing Prices")

# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')

# Set the x-axis label
plt.xlabel('Size (1000 sqft)')

plt.legend()
plt.show()
```
